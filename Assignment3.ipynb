```python
merged2_df['pdi'] = pd.to_numeric(merged2_df['pdi'], errors='coerce')
merged2_df['pdi'].fillna(merged2_df['pdi'].mean(), inplace=True)
merged2_df['mas'] = pd.to_numeric(merged2_df['mas'], errors='coerce')
merged2_df['mas'].fillna(merged2_df['mas'].mean(), inplace=True)
merged2_df['uai'] = pd.to_numeric(merged2_df['uai'], errors='coerce')
merged2_df['uai'].fillna(merged2_df['uai'].mean(), inplace=True)
merged2_df['idv'] = pd.to_numeric(merged2_df['idv'], errors='coerce')
merged2_df['idv'].fillna(merged2_df['idv'].mean(), inplace=True)
```

```python
grouped_df = merged2_df.groupby(['Team', 'season'])
print(grouped_df.groups)
```

```python
def coefficient_of_variation(data):
    mean_value = np.mean(data)
    std_dev = np.std(data)
    return (std_dev / mean_value) * 100 if mean_value != 0 else np.nan

def cultural_heteroheneity(data):
    n = len(data)
    s = 0
    unique_countries = data.unique()
    for country in unique_countries:
        i = data[data == country].count()
        pi = i / n
        s += pi * np.log(pi) * -1
    return s / (-1 * np.log(1/n))

def cultural_distance(*data):
    n = len(data[0])
    s = 0
    for entry in data:
        for i in entry:
            for j in entry:
                s += (i - j)**2
    return np.sqrt(s / (n * (n-1) * 4))
            

def collectivism(data):
    s = sum(data)
    n = len(data)
    return 100 - (s / n)

agg_dict = {
    'age': [('range',np.ptp), 'median', ('iqr',st.iqr), 'mean', 'std', ('cv',coefficient_of_variation)],
    'player_weight': [('range',np.ptp), 'median', ('iqr',st.iqr), 'mean', 'std', ('cv',coefficient_of_variation)],
    'player_height': [('range',np.ptp), 'median', ('iqr',st.iqr), 'mean', 'std', ('cv',coefficient_of_variation)],
    'country': [('cultural_heteroheneity', cultural_heteroheneity)],
    'idv': [('collectivism', collectivism)]
}

result_df = grouped_df.agg(agg_dict)
result_df['cultural_distance'] = grouped_df.apply(lambda x: pd.Series({'cultural_distance': cultural_distance(x['pdi'],x['idv'],x['mas'],x['uai'])}))

result_df.head()
```
